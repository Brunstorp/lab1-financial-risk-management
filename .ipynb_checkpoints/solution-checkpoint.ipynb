```python
#imports
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the Excel file
file_path = "DataLab1.xlsx" 

# Read the first two columns (A and B)
df = pd.read_excel(file_path, usecols=[0, 1])

# Display the first few rows (for testing)
df.head()
```

```python
#Here we create a Profit loss dataframe
PL = df["PL"]

#initialize alpha (confidence level)
alpha = 0.01
```

```python
VAR_list = []
cutoff = int(500*alpha - 1)
for i in range(500,len(PL)):
    window = PL.iloc[i-500:i]
    sorted_window = window.sort_values()
    VAR_list.append(-sorted_window.iloc[cutoff])
#print(VAR_list)
print(len(VAR_list))
```

```python
plt.plot(VAR_list)
```

```python
# 2.a
mu = df["PL"].iloc[0:500].mean()
```

```python
print(mu)
```

```python
df["Unexpected Losses"] = df["PL"]-mu
unexpected_losses = df["Unexpected Losses"]
```

```python
print(df["Unexpected Losses"])
```

```python
# 2.b
l = 0.94
e2_tm1 = 0 # e2_0
s2_tm1 = unexpected_losses.iloc[0:500].var() #sigma2_0
#print(s2_0)
sigmas = []
for t in range(len(unexpected_losses)):
    s2_t = (1-l)*e2_tm1 + l*s2_tm1
    sigmas.append(s2_t)
    s2_tm1 = s2_t
    e2_tm1 = unexpected_losses.iloc[t]**2
    
```

```python
print(sigmas)
```

```python
plt.plot(sigmas)
```

```python
# 2.c

EWMA_conditional_variances = np.sqrt(sigmas)
print(EWMA_conditional_variances)
plt.plot(EWMA_conditional_variances)
```
